```python
import matplotlib.pyplot as plt
import plotly
import seaborn
```

```python
index = open('房价指数.csv')
```

```python
data = pd.read_csv(index)
```

```python
data
```

```python
date = data['Year/Month']
```

```python
date
```

```python
index = data['Price Index']
```

```python
index
```

```python
fig = plt.figure(figsize=(15,6))
```

```python
plt.plot(date,
         index,
         linestyle = '-',
         linewidth = 1,
         color = 'brown',
         marker = '*',
         markersize = 2,
         label = 'price index'
        )

plt.xlabel('Year')
plt.ylabel('Price index')

date = date.sort_index()
plt.plot(date.index, index.values)
plt.xticks([1, 13, 25, 37, 49, 61,73,85,97,109,121],['2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018'])
```
