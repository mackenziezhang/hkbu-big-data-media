```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from matplotlib.ticker import MultipleLocator, FormatStrFormatter  
import seaborn as sns
import csv
import numpy as np
```

```python
df = pd.read_csv('zhilian-csv.csv')
```

```python
df
```

```python
def split_salary(i):
    try:
        splt = i.split('-',1)
        first = float(splt[0])
        second = float(splt[1])
        return int((first+second)/2)
    except:
        return float(i)
df['月薪平均'] = df['职位月薪'].apply(split_salary)
#由于原本的“职位月薪”为字符串表示的数量范围，难以进行统计，所以取中间值变成一个数字。
```

```python
#查看总体薪酬情况
font = FontProperties(fname = '/Users/mackenziezhang/library/Fonts/SourceHanSansCN-Medium.otf', size = 12)
div_1000 = df['月薪平均']/1000
ax1 = plt.figure(figsize=(10, 6)).add_subplot(111)
sns.distplot(div_1000, kde=False, bins=30)
ax1.set_title('薪酬分布', Fontproperties = font)
ax1.set_xlabel('薪酬（K/月）', Fontproperties = font)
ax1.set_xticks(range(5, 60, 5))
plt.show()
```
